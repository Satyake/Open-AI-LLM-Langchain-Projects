```python
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader, ServiceContext
from llama_index.core.prompts.prompts import SimpleInputPrompt
```

```python
!mkdir Data
```

```python
loader=SimpleDirectoryReader('/content/Data')
```

```python
documents=loader.load_data()
```

```python
#documents
!pip -q install llama-index-embeddings-huggingface llama-index-embeddings-instructor
```

```python
#load embedding model
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
embed_model=HuggingFaceEmbedding(model_name='sentence-transformers/all-mpnet-base-v2')
```

```python
from llama_index.core import PromptTemplate
```

```python
system_prompt="""
<|SYSTEM|># you are a QA Assistant. Your goal is to answer as accurately as
possible based on the instructions and context provided

"""
```

```python
query_wrapper_prompt=PromptTemplate("<|USER|>{query_str}<|ASSISTANT|>")
```

```python
#system
#assistant
#user
```

```python
!pip -q install llama-index-llms-huggingface
```

```python
from llama_index.llms.huggingface import HuggingFaceLLM
```

```python
import torch
llm=HuggingFaceLLM(
    context_window=4096,
    max_new_tokens=256,
    generate_kwargs={'temperature':0.7, "do_sample": False},
    system_prompt=system_prompt,
    query_wrapper_prompt=query_wrapper_prompt,
    tokenizer_name='mistralai/Mistral-7B-Instruct-v0.1',
    model_name='mistralai/Mistral-7B-Instruct-v0.1',
    device_map='auto',
    stopping_ids=[50278, 50279, 1, 0],
    tokenizer_kwargs={'max_length':4096},
    model_kwargs={'torch_dtype':torch.float16}
)
```

```python
from llama_index.core import ServiceContext
ServiceContext=ServiceContext.from_defaults(
    chunk_size=1024, llm=llm, embed_model=embed_model
)
```

```python
from llama_index.core import VectorStoreIndex
```

```python
index=VectorStoreIndex.from_documents(documents,service_context=ServiceContext)
```

```python
query_engine=index.as_query_engine()
```

```python
result=query_engine.query("what is attention mechanism?")
```
